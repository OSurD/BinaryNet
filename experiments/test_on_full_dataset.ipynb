```python
from BinaryNet.utils.ClassificationTrainer import ClassificationTrainer
from BinaryNet.utils.Dataset import SCWithFeatures, FilteredSubsetSC
from BinaryNet.utils.SoundUtils import prepareMelSpectrogram

from BinaryNet.models.SimpleCNN import SimpleCNN

import os
```

```python
DATA_DIR = os.path.abspath('../data')
SPEECHCOMMANDS_DATASET_PATH = os.path.join(DATA_DIR, "SPEECHCOMMANDS")

train = SCWithFeatures(FilteredSubsetSC(SPEECHCOMMANDS_DATASET_PATH, "training"), prepareMelSpectrogram)
val   = SCWithFeatures(FilteredSubsetSC(SPEECHCOMMANDS_DATASET_PATH, "validation"), prepareMelSpectrogram)
class_counts = len(train.base.target)
```

```python
trainer = ClassificationTrainer(train, val)
```

```python
SIMPLECNN_DIR = os.path.abspath('./models/SimpleCNN')
SIMPLECNN_CHECKPOINTS_DIR = os.path.join(SIMPLECNN_DIR, '/checkpoints_full')
os.makedirs(SIMPLECNN_CHECKPOINTS_DIR, exist_ok=True)

model = trainer.fit(20, SimpleCNN(class_counts), bs_test=1000, save_path=SIMPLECNN_CHECKPOINTS_DIR, save_interim_models=True)
```

```python
trainer.visualize(filename=os.path.join(SIMPLECNN_DIR, "learning.jpg"))
```

```python
from BinaryNet.models.XNORSimpleCNN import XNORSimpleCNN

SIMPLECNN_DIR = os.path.abspath('./models/XNORSimpleCNN')
SIMPLECNN_CHECKPOINTS_DIR = os.path.join(SIMPLECNN_DIR, 'checkpoints_full')
os.makedirs(SIMPLECNN_CHECKPOINTS_DIR, exist_ok=True)

model = trainer.fit(20, XNORSimpleCNN(class_counts), bs_test=1000, save_path=SIMPLECNN_CHECKPOINTS_DIR, save_interim_models=True)
```
